Clasificación de variedades de flores iris.

```python
#0. Importando dataset.
import pandas as pd
import numpy as np

#Cargando datos pima indians dataset
dataset = pd.read_csv(r'/content/Iris_dataset.csv')
print(dataset)
dataset = np.array(dataset)
```

```python
#1. Dividing dataset into input (X) and output (Y) variables
X = dataset[:,0:4]
Y = dataset[:,4]
```

```python
#2. Discretizing the categorical data.
Y2=np.zeros((len(Y)))
for i in range(0,len(Y)):
  if Y[i]=='Iris-setosa':
     Y2[i]=1
  else:
     if Y[i]=='Iris-versicolor':
        Y2[i]=2
     else:
        Y2[i]=3
```

```python
#3. Split input and output into training and testing sets
import sklearn
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train, Y_test= train_test_split(X,Y2,test_size=0.2,random_state=14541)
```

```python
#4. Data normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

===================================================================

```python
#5. Evaluando casos mediante KNN
from sklearn.neighbors import KNeighborsClassifier
Modelo_0 = KNeighborsClassifier(3)
Modelo_0.fit(X_train, Y_train)
Y_pred_0 =Modelo_0.predict (X_test)
# Graficando resultados
print('Mostrando el Expected y predicted')
for j in range(0,len(Y_test)): 
  print (Y_test[j],' ',Y_pred_0[j])
print('')
```

```python
#5_B. Calculating confusion matrix
import sklearn
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(Y_test, Y_pred_0)
print(matrix)
print('')
```

```python
#5_C. calculando desempeño (Exactitud)
from sklearn.metrics import accuracy_score,precision_score
print("Accuracy ",accuracy_score(Y_test, Y_pred_0))
```

```python
#5_D: Creating a new target
X_new = np.array([[3, 2, 1, 0.2], [  4.9, 2.2, 3.8, 1.1 ], [  5.3, 2.5, 4.6, 1.9 ]])
X_new = scaler.transform(X_new)
#Prediction of the species from the input vector
prediction_0 = Modelo_0.predict(X_new)
print("Predictions of Species are: ",prediction_0)
```
