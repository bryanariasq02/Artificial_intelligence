```python
# 0. Importando el conjunto de datos
import pandas as pd
import numpy as np

# Cargando el dataset de la calidad de los vinos
dataset = pd.read_csv(r'/content/drogas.csv')
dataset.info()
print("==========================================")
dataset["Sex"] = dataset["Sex"].replace("F", "1")
dataset["Sex"] = dataset["Sex"].replace("M", "2")
dataset["Sex"] = dataset["Sex"].astype(np.float16)

dataset["BP"] = dataset["BP"].replace("HIGH", 1)
dataset["BP"] = dataset["BP"].replace("LOW", 2)
dataset["BP"] = dataset["BP"].replace("NORMAL", 3)
dataset["BP"] = dataset["BP"].astype(np.float16)

dataset["Cholesterol"] = dataset["Cholesterol"].replace("HIGH", 1)
dataset["Cholesterol"] = dataset["Cholesterol"].replace("LOW", 2)
dataset["Cholesterol"] = dataset["Cholesterol"].replace("NORMAL", 3)
dataset["Cholesterol"] = dataset["Cholesterol"].astype(np.float16)

dataset = np.array(dataset)
dataset.shape
```

```python
#1. Diviendo el dataset en conjuntos de entrada(X) y salida(Y)
X = dataset[:,0:5]
Y = dataset[:,5]
```

```python
#2. Discreción categorica de los datos de salida
Y2 = np.zeros((len(Y)))
for i in range(0, len(Y)):
  if Y[i]=='drugA':
     Y2[i] = 1
  elif Y[i] == 'DrugY':
     Y2[i] = 2
  elif Y[i] == 'drugX':
     Y2[i] = 3
  elif Y[i] == 'drugA':
     Y2[i] = 4
  elif Y[i] == 'drugB':
     Y2[i] = 5
  else:
     Y2[i] = 6

print(f"Tipos de droga: \n {Y}")
print("")
print(f"Tipos de droga ya categorizadas: \n{Y2}")
```

```python
#3. Split input and output into training and testing sets
import sklearn
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train, Y_test= train_test_split(X,Y2,test_size=0.2,random_state=14541)
```

```python
#4. Data normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#5. Evaluando casos mediante KNN
from sklearn.neighbors import KNeighborsClassifier
Modelo_0 = KNeighborsClassifier(3)
Modelo_0.fit(X_train, Y_train)
Y_pred_0 =Modelo_0.predict (X_test)
# Graficando resultados
print('Mostrando el Expected y predicted')
for j in range(0,len(Y_test)): 
  print (Y_test[j],' ',Y_pred_0[j])
print('')
```

```python
#5_B. Calculating confusion matrix
import sklearn
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(Y_test, Y_pred_0)
print(matrix)
print('')
```

```python
#5_C. calculando desempeño (Exactitud)
from sklearn.metrics import accuracy_score,precision_score
print("Accuracy ",accuracy_score(Y_test, Y_pred_0))
```
